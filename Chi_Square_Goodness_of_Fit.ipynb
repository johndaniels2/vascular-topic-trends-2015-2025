```python
import pandas as pd
import numpy as np
from scipy.stats import chisquare
```

```python
data = {
    'Aortic':               [1007, 241, 217, 174, 130, 441],
    'PAD':                  [ 806, 189, 152, 142, 133, 228],
    'Cerebrovascular':      [ 370,  70,  32,  34,  52, 138],
    'Venous/Lymphatic':     [ 361,  47,  25,  33,  46, 136],
    'Dialysis Access':      [ 206,  36,  14,  28,  35,  96],
    'Practice Management':  [ 488,  49,  22,  21,  29, 153],
    'Imaging':              [ 107,  20,  20,  12,   9,  22],
    'Compression Syndrome': [  76,  14,   5,  12,   8,  18],
}
societies = ['SVS', 'SAVS', 'CSVS', 'WVS', 'NESVS', 'SCVS']
df = pd.DataFrame(data, index=societies)
df
```

```python
clinical = ['Aortic', 'PAD', 'Cerebrovascular', 'Venous/Lymphatic',
            'Dialysis Access', 'Imaging', 'Compression Syndrome']

observed = df[clinical].sum()
N = observed.sum()
print(observed.to_string())
print(f'\nTotal presentations (clinical domains): {int(N)}')
```

```python
acgme = pd.Series({'Aortic': 9.5, 'PAD': 38.5, 'Cerebrovascular': 8.7,
                   'Venous/Lymphatic': 5.9, 'Dialysis Access': 9.2,
                   'Imaging': 19.2, 'Compression Syndrome': 0.7})
vsite = pd.Series({'Aortic': 23.9, 'PAD': 26.1, 'Cerebrovascular': 17.4,
                   'Venous/Lymphatic': 19.6, 'Dialysis Access': 4.3,
                   'Imaging': 4.3, 'Compression Syndrome': 4.3})

acgme = acgme[clinical] / acgme.sum()
vsite = vsite[clinical] / vsite.sum()
```

```python
expected = acgme * N
chi2, p = chisquare(f_obs=observed.values, f_exp=expected.values)
print('Conference vs ACGME operative volume')
print(f'Chi-square statistic: {chi2:.2f}')
print(f'Degrees of freedom: {len(observed) - 1}')
print(f'P-value: {p:.4g}')
```

```python
expected = vsite * N
chi2, p = chisquare(f_obs=observed.values, f_exp=expected.values)
print('Conference vs VSITE exam blueprint')
print(f'Chi-square statistic: {chi2:.2f}')
print(f'Degrees of freedom: {len(observed) - 1}')
print(f'P-value: {p:.4g}')
```
